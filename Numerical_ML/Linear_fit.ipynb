```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
mpl.rcParams['figure.dpi'] = 350
mpl.rcParams['axes.grid'] = True
mpl.rcParams['axes.axisbelow'] = True
mpl.rcParams['figure.facecolor'] = 'white'
mpl.rcParams['savefig.facecolor'] = 'white'
mpl.rcParams['axes.formatter.useoffset'] = False
```

```python
def f(x, w=0.3, b=0.1):

    return w * x + b
```

```python
# Generating Data

x_input = np.random.uniform(-10, 10, 1000)
y = f(x_input)  + 5*np.random.normal(0, 1, x_input.size) # The target
```

```python
# Initializing variable

w = np.random.uniform(-10, 10)
b = np.random.uniform(-10, 10)
```

# Before training

```python
plt.title(f'w: {round(w, 3)}    b: {round(b, 3)}')
plt.plot(x_input, f(x_input), lw=3, color = 'orange', ls = '--', label='Target function')
plt.scatter(x_input, y, s = 2)
plt.plot(x_input, f(x_input, w, b), lw=2.5, color = 'purple', ls=':', label='NN model')
plt.legend()
plt.show()
plt.close()
```

```python
# lr = 0.1 # Expload case
lr = 0.001 # Good case

def mse(
    x_input,
    y_hat, target, # Model & target output
    w, b # Tunable variable
):

    return (y_hat(x_input, w, b) - y)**2 / 2
```

# Training

```python
iteration = 5000
for i in range(iteration + 1):
    div_w = (mse(x_input, f, y, w+0.05, b) - mse(x_input, f, y, w-0.05, b))/0.1
    div_b = (mse(x_input, f, y, w, b+0.05) - mse(x_input, f, y, w, b-0.05))/0.1

    w -= lr*div_w.mean()
    b -= lr*div_b.mean()

    if i % int(iteration/10)  == 0:
        print(f'{i}th: W={round(w,  5)} b={round(b, 5)}')
```

# After training

```python
plt.title(f'w: {round(w, 3)}    b: {round(b, 3)}')
plt.plot(x_input, f(x_input), lw=3, color = 'orange', ls = '--', label='Target function')
plt.scatter(x_input, y, s = 2)
plt.plot(x_input, f(x_input, w, b), lw=2.5, color = 'purple', ls=':', label='NN model')
plt.legend()
plt.show()
plt.close()
```
